```python
from langchain.embeddings import HuggingFaceEmbeddings
embedding_model_name = "sentence-transformers/all-MiniLM-L6-v2"
embeddings = HuggingFaceEmbeddings(model_name=embedding_model_name)
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
import os

# Specify the path to the folder containing PDF files
pdf_folder = "/content/drive/MyDrive/IndianCuisine"  # Update with your folder path

# List all PDF files in the folder
pdf_files = [os.path.join(pdf_folder, file) for file in os.listdir(pdf_folder) if file.endswith(".pdf")]

# List to store the loaded data
all_data = []

# Loop through each file and load the data
for pdf_file in pdf_files:
    loader = PyMuPDFLoader(pdf_file)
    data = loader.load()
    all_data.extend(data)

# Now all_data contains the data from all the PDFs
for doc in all_data:
    print(doc)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(all_data)
```

```python
 docs
```

```python
vector_db = Weaviate.from_documents(
    docs, embeddings, client=client, by_text=False
)
```
